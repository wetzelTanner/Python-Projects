## Section 1. Import Pandas package and dataset

### Import the necessary libraries

```python
import pandas as pd
```

### Read the given dataset into a Pandas DataFrames type. Name the DataFrames as `data`

Turns out that `read_csv()` function can also interpret `csv` files hosted on the Internet. We just need to supply the URL as a string to the function instead of supplying the path to find the file.

```python
data_url = 'https://raw.githubusercontent.com/BlueJayADAL/CS121/master/datasets/iris.csv'
```

```python
data = pd.read_csv(data_url, header = None)

```

## Section 2: Sneak peek to the DataFrame

### Check out the contents in the DataFrame `data` using `head()` method.

```python
# head() by default displays the first 5 records of data from the DataFrame

data.head()
```

### Change the column names to `sepal_length`, `sepal_width`, `petal_length`, `petal_width`, and `species` respectively. 

```python
data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
data.head()
```

### Check out the basic information of the DataFrame including column names, dtypes and memory usage, etc.


```python
data.info()
```

### Check the existence of missing data.

```python
data.isnull().sum()
```

### Show the descriptive statistics of all numeric columns in the DataFrame

```python
data.describe()
```

## Section 3: Line Plot and Area Plot

### Show a line graph of all the variables in the DataFrame

```python
data.plot(kind = 'line')
```

### Show a line graph with only `sepal_length` and `sepal_width` variables from the DataFrame

```python
data[['sepal_length', 'sepal_width']].plot(kind = 'line', 
                                           title = 'Line graph of sepal length and width', 
                                           xlabel = 'Flower instances', 
                                           ylabel = 'Meaurement in cm')
```

### Show an area graph of all the variables in the DataFrame

```python
data.plot(kind = 'area', alpha = 0.6)
```

## Section 4: Bar Plot

### Create a bar graph with `sepal_length` for all data entries. Enlarge the graph to show details.

```python
data[['sepal_length']].plot(kind = 'bar', figsize = (12, 5))
```

### Plot a bar graph with the count of each possible `sepal_length`

```python
# Step 1: select only column 'sepal_length' and then count all unique values

sepal_len_counts = data['sepal_length'].value_counts()

sepal_len_counts
```

```python
# Step 2: Sort the result from step 1 with index

sepal_len_counts_sorted = sepal_len_counts.sort_index()

sepal_len_counts_sorted
```

```python
# Step 3: plot the result from step 2 with a bar graph

sepal_len_counts_sorted.plot(kind = 'bar', figsize = (10, 5))
```

### Plot a bar graph with the average `sepal_length` for each flower `species` type

```python
sepal_len_by_species = data.groupby(by = 'species').mean()['sepal_length']

sepal_len_by_species.head()
```

```python

sepal_len_by_species.plot(kind = 'bar', title = 'Average sepal length for each species', xlabel = 'Flower instances', ylabel = 'Meaurement in cm')
```

### Show the figure above in horizontal orientation.

```python

sepal_len_by_species.plot(kind = 'barh', title = 'Average sepal length for each species', xlabel = 'Flower instances', ylabel = 'Meaurement in cm' )

```

## Section 5: Histogram

### Display a histogram of `sepal_length` with all data entries.

```python

data['sepal_length'].plot(kind = 'hist', bins = 30, color = 'orange')
```

### Display histograms of all four variables in subplots

```python
# Not very visible as they all stack together.

data.plot(kind = 'hist', bins = 30)
```

```python
# Split these four histograms into their own subplot

data.plot(kind = 'hist', subplots = True, layout = (2,2), bins = 30, figsize = (10,10))
```

## Section 6: Scatter Plot
### Show a simple scatterplot with two variables `sepal_length` and `sepal_width`.

```python
data.plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width')

```

### Show a scatterplot with two variables `sepal_length` and `sepal_width`, and color-code each data point with its `petal_length`

```python
data.plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width', c = 'petal_length', cmap = 'inferno')

```

### Show a scatterplot with two variables `sepal_length` and `sepal_width`, and color-code each data point with its `species`

```python
# Check to see all different flower species

data['species'].unique()
```

```python
# Step 1: Define a function to be used to change 'species' column
# If flower is 'Iris-setosa', convert to 0
# If flower is 'Iris-versicolor', convert to 1
# If flower is 'Iris-virginica', convert to 2

def conv_species(in_str):
    if in_str == 'Iris-setosa':
        return 0
    elif in_str == 'Iris-versicolor':
        return 1
    elif in_str == 'Iris-virginica':
        return 2
    
    
```

```python
# Step 2: Change the species column from string to numbers

data['species'] = data['species'].apply(conv_species)


data.head()
```

```python
data['species'].value_counts()
```

```python
# Step 3: show the scatter plot and use `species` to color code

data.plot(kind = 'scatter', x = 'sepal_length', y = 'sepal_width', c = 'species', cmap = 'Accent')
```
